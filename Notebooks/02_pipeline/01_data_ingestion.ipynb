```python
%sql
USE CATALOG ecommerce;
USE SCHEMA bronze;
```

SOURCE 1 

```python
df_orders = spark.read.format("csv") \
.option("header", True) \
.option("inferSchema", True) \
.load("abfss://projectstorage@eprojectstorageacc.dfs.core.windows.net/bronze/orders/Ecommerce.csv")

df_orders.limit(10).display()
```

Save as Bronze Table

```python
df_orders.write.format("delta") \
.mode("overwrite") \
.saveAsTable("orders")
```
